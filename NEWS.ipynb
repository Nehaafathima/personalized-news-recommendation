```python
# Step 1: Install necessary libraries (if not already installed)
!pip install nltk scikit-learn pandas

# Step 2: Import Libraries
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

# Step 3: Sample News Dataset (You can replace this with your own dataset)
data = {
    'title': [
        'Stock Market Hits Record Highs',
        'Local Elections Results Announced',
        'Scientists Discover New Planet',
        'Tech Companies Release Earnings Reports',
        'Global Warming Impact on Wildlife',
        'AI Revolution in Healthcare',
        'Bitcoin Prices Surge Again',
        'Education Reforms Introduced in Schools',
        'Sports Update: Football Finals',
        'Medical Breakthrough in Cancer Research'
    ],
    'content': [
        'The stock market reached new heights today as investors showed strong confidence...',
        'The local government announced the election results with surprising outcomes...',
        'A new planet was discovered by NASA scientists far beyond our solar system...',
        'Apple, Google and other tech giants released their quarterly earnings today...',
        'Wildlife is being affected significantly due to climate change and global warming...',
        'Artificial Intelligence is making big advances in diagnosing health conditions...',
        'Bitcoin continues to rise as investors see it as a store of value...',
        'New education policies have been introduced to improve learning standards...',
        'The football finals saw an intense match between top clubs...',
        'Researchers found a new treatment method with promising cancer results...'
    ]
}

df = pd.DataFrame(data)

# Step 4: Preprocessing - TF-IDF Vectorization
stop_words = stopwords.words('english')
tfidf = TfidfVectorizer(stop_words=stop_words)
tfidf_matrix = tfidf.fit_transform(df['content'])

# Step 5: Compute Cosine Similarity
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

# Step 6: Recommendation Function
def get_recommendations(title, cosine_sim=cosine_sim):
    idx = df[df['title'] == title].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:6]  # Get top 5 similar news
    news_indices = [i[0] for i in sim_scores]
    return df['title'].iloc[news_indices]

# Step 7: Test the Recommendation System
user_interest = 'AI Revolution in Healthcare'
print(f"News similar to '{user_interest}':\n")
recommended = get_recommendations(user_interest)
for i, news in enumerate(recommended, 1):
    print(f"{i}. {news}")
```
